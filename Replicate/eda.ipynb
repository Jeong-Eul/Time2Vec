```python
import pandas as pd
import numpy as np
```

```python
med_concept = pd.read_csv('infusionDrug.csv')
```

```python
vasoactive_inotropic = ['Norepinephrine (mcg/min)', 'Dopamine (mcg/min)', 'Phenylephrine (mcg/min)', 'Epinephrine (mcg/min)',
                         'Nitroprusside (mcg/kg/min)', 'Dobutamine (mcg/kg/min)']
```

```python
interest = med_concept.drugname.isin(vasoactive_inotropic)
obj = med_concept[interest][['patientunitstayid', 'infusionoffset','drugname', 'infusionrate']]
obj
```

```python
target = obj.pivot_table(index = ['infusionoffset','patientunitstayid'], columns = 'drugname')
```

```python
target.to_csv('after_pivot.csv')
```

```python
df = pd.read_csv('after_pivot.csv')
df
```

```python
df = df[df['infusionoffset']>0].sort_values(by=['patientunitstayid', 'infusionoffset'])
df
```

```python
df['vasoactive/inotropic']=df.iloc[:,2:].sum(axis = 1).values
df
```

```python
sample = df[['patientunitstayid','infusionoffset','vasoactive/inotropic']]
sample
```

```python
sample[sample['vasoactive/inotropic']==0.0]
```

```python
sample['vasoactive/inotropic'] = sample['vasoactive/inotropic'].apply(lambda x: 1 if x > 0 else 0)
sample
```

```python
sample.reset_index(drop=True, inplace = True)
sample.to_csv('vasoactive_inotropic.csv')
```

```python
sample['vasoactive/inotropic'].value_counts()
```

```python
import time
import random
def check_class_ratio(dataset):
    class_ratio = round(np.mean(dataset['vasoactive/inotropic']), 2)
    return class_ratio



def random_split_patient(df, train_ratio, Threshold, n_trial):
    data = df.copy()
    
    search_time = time.time()
    
    for T in range(n_trial):
        start = time.time()
        array = data.patientunitstayid.unique()
        
        seed = random.randint(1, 10000)
        np.random.seed(seed) 
        np.random.shuffle(array)


        split_point = int(train_ratio * len(array))
        patient_for_train, patient_for_test = np.split(array, [split_point])

        

        condition_train = data.patientunitstayid.isin(patient_for_train)
        holdout_train = data[condition_train]

        condition_test = data.patientunitstayid.isin(patient_for_test)
        holdout_test = data[condition_test]
        
        holdout_test = holdout_test.sort_values(by=['patientunitstayid', 'infusionoffset'])
                 

        train_class_ratio  = check_class_ratio(holdout_train)
        test_class_ratio  = check_class_ratio(holdout_test)
                
            
        if (train_class_ratio - test_class_ratio) >= Threshold:
            
            break  # 클래스 비율이 모두 threshold 이상일 경우 반복문 종료
        
        if T % 100 == 0:
            print('Trial: ', T)
        
        if T == 9999:
            print('최대 Trial 달성, 분할 불가')
        
        end = time.time()
#         print('Trial 종료, 소요시간:{}'.format(end-start))
        
    train = holdout_train.drop(['patientunitstayid'], axis = 1)
    test = holdout_test.drop(['patientunitstayid'], axis = 1)
    search_time_end = time.time()
    
    print('train set : test set = {} : {}'.format(train_ratio, 1-train_ratio))
    print('Train set class: ', train['vasoactive/inotropic'].value_counts())
    print('Test set class: ', test['vasoactive/inotropic'].value_counts())
    print('-'*20)
    print('Train class ratio: {}:{}'.format((train['vasoactive/inotropic'].value_counts()[0])/len(train), (train['vasoactive/inotropic'].value_counts()[1])/len(train)))
    print('Test class ratio: {}:{}'.format((test['vasoactive/inotropic'].value_counts()[0])/len(test), (test['vasoactive/inotropic'].value_counts()[1])/len(test)))
    print('Split seed: ',seed)
    print('train ratio:', train_ratio)
    print('Threshold:', Threshold)
    print('총 소요 시간(초):{}'.format(search_time_end - search_time))
    print('시도한 trial 수: ', T)
    
    return train, test

trainset, testset = random_split_patient(sample, train_ratio=0.8, Threshold=0.01, n_trial=200)
```

```python
trainset.to_csv('trainset.csv')
testset.to_csv('testset.csv')
```
